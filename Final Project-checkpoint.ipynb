### Key Facts About Alcohol Consumption:
* The harmful use of alcohol is a causal factor in more than 200 disease and injury conditions.
* Worldwide, 3 million deaths every year result from harmful use of alcohol. This represents 5.3% of all deaths.
* Overall, 5.1% of the global burden of disease and injury is attributable to alcohol, as measured in disability-adjusted life years (DALYs).
* Beyond health consequences, the harmful use of alcohol brings significant social and economic losses to individuals and society at large.
* Alcohol consumption causes death and disability relatively early in life. In people aged 20–39 years, approximately 13.5% of total deaths are attributable to alcohol.
* There is a causal relationship between harmful use of alcohol and a range of mental and behavioural disorders, other noncommunicable conditions and injuries.

### Key Facts About Road Accident Deaths:
* The United Nations General Assembly has set an ambitious target of halving the global number of deaths and injuries from road traffic crashes by 2030 (A/RES/74/299).
* Road traffic injuries are the leading cause of death for children and young adults aged 5-29 years.
* Approximately 1.3 million people die each year as a result of road traffic crashes.
* More than half of all road traffic deaths are among vulnerable road users: pedestrians, cyclists, and motorcyclists.
* 93% of the world's fatalities on the roads occur in low- and middle-income countries, even though these countries have approximately 60% of the world's vehicles.
* Road traffic crashes cost most countries 3% of their gross domestic product.

### Step2: Data Preprocessing - Data preprocessing is defined as removal of noisy and irrelevant data from collection. 
* Cleaning in case of Missing values.
* Cleaning noisy data, where noise is a random or variance error.
* Cleaning with Data discrepancy detection and Data transformation tools.

```python
# cleaning the datasets 
def cleanValueColumn(values):
    new_values = []
    for value in values:
        index = 0
        for letter in value:
            if letter=="[":
                new_values.append(float((value[:index-1]).strip()))
            index = index + 1
    return new_values
```

```python
values_alc = list(df_alcohol_consumption['Value'])
values_road = list(df_road_accident['Value'])

cleaned_values_alc = cleanValueColumn(values_alc)
cleaned_values_road = cleanValueColumn(values_road)

df_alcohol_consumption["Value"] = cleaned_values_alc
df_road_accident["Value"] = cleaned_values_road 
```

```python
# Creating subset of the datasets with relevant columns and dropping the irrelevant ones
def subsetCreation(df):
    df = df[['ParentLocation','Location', 'Period', 'FactValueNumericLow','FactValueNumericHigh','Value']]
    return df
```

```python
df_alch_cons = subsetCreation(df_alcohol_consumption)
df_road_dth = subsetCreation(df_road_accident)
```

```python
# taking the average of each year which was appearing thrice so that the ocurrance of each year for each country should be presented uniquely
def alcDf(df):
    dict_alch = []
    location_arr = list(df["Location"].unique())
    period_arr = list(df["Period"].unique())
    year_index = 0
    
    for x in location_arr:
        for y in period_arr:
            element = df[(df["Location"]==x) & (df["Period"]==y)]
            try:
                col_one_two = list(element[['ParentLocation','Location']].values[0])
                col_one_two.append(int(y))
                values = list(element.mean()[1:])
                element_val = col_one_two + values 
                dict_alch.append(element_val)
            except:
                pass
    
    df = pd.DataFrame(dict_alch,columns=list(df.columns))
    
    return df
```

```python
df_alch_cons = alcDf(df_alch_cons) #calling the dataset
```

```python
# user define function for taking the subset of the whole dataset according to the selected countries
def contrySubset(country,df):
    df = df[df["Location"]==country]
    return df
```

```python
# user define function for concatenating the two cleaned subset 
def joinDf(df_arr):
    df = pd.DataFrame()
    for x in df_arr:
        df=pd.concat([df,x],ignore_index =True)
    return df
```

```python
def plotGraph(df1, df2, variable, variable1, country1,country2):
 
 palette = sns.color_palette()
 c1 = random.choice(palette)
 idx = palette.index(c1)
 palette.pop(idx)
 c2 = random.choice(palette)
 
 plt.figure(figsize=(16, 16))
 plt.subplot(2,2,1)
 plt.title(str(variable + " (DEATHS PER 100,000 OF POPULATION)"))
 plot_one = sns.lineplot(data = df2, x = 'Period', y = df2[df2["Location"]==country1][variable]
                         , color = c1 , label = country1)
 plot_two = sns.lineplot(data = df2, x = 'Period', y = df2[df2["Location"]==country2][variable],
                         color = c2, label = country2)
 
 plt.subplot(2,2,2)
 plt.title(str(variable1 + " (Consumption per capita)"))
 plot_one = sns.lineplot(data = df1, x = 'Period', y = df1[df1["Location"]==country1][variable1], 
                         color = c1 , label = country1)
 plot_two = sns.lineplot(data = df1, x = 'Period', y = df1[df1["Location"]==country2][variable1], 
                         color = c2, label = country2)

 plt.subplot(2,2,3)
 plt.title("Average Deaths per 100,000 from 2000-2019")
 plt.bar(x=[country1,country2],height=[df2[df2["Location"]==country1][variable].mean(),
 df2[df2["Location"]==country2][variable].mean()],color=[c1,c2])
 
 plt.subplot(2,2,4)
 plt.title("Average Consumption per capita from 2000-2019")
 plt.bar(x=[country1,country2],height=[df1[df1["Location"]==country1][variable1].mean(),
 df1[df1["Location"]==country2][variable1].mean()],color=[c1,c2])
 plt.show()
```
